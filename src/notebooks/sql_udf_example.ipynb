# UDF

```python
%%sql
CREATE EXTENSION if not exists plpgsql;
CREATE EXTENSION if not exists plpythonu;
```

```python
%%sql
DROP FUNCTION IF EXISTS mult1(v int);
create function mult1(v int) returns int as $$
SELECT v * v * 100;
$$ language sql;
```

```python
%%sql
SELECT age, age*age, mult1(age) FROM sailors;
```

```python
%%sql
DROP TABLE IF EXISTS blah;
create table blah (qty int);

drop function if exists plpsqlf(v int);
create function plpsqlf(v int) returns text as $$
DECLARE
  qty int = 10;
BEGIN
  if (qty * v > 10) THEN
    qty = -10;
  end if;
  insert into blah values(qty);
  return null;
END;
$$ language plpgsql;
```

```python
%%sql
select * from blah;
```

```python
%%sql
DROP FUNCTION IF EXISTS search(word text);
CREATE FUNCTION search(word text) returns text as $$
import requests
r = requests.get("https://en.m.wikipedia.org/w/index.php?title=Special:Search&search=%s" % word)
return r.content.decode('unicode-escape')
$$ language plpythonu;
```

```python
%%sql
drop table if exists unis;
create table unis(name text);
insert into unis values ('columbia university'), ('uc berkeley'), ('harvard');
```

```python
rows = %sql SELECT search(name) from unis;
```
